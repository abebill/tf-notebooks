```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = tf.placeholder(tf.int32, shape=[2], name='my_input')
b = tf.reduce_prod(a, name='prod_a')
c = tf.reduce_sum(a, name='sum_b')
d = tf.add(b, c, name='output')

sess = tf.Session()
input_dict = {a: np.array([5, 3], dtype=np.int32)}
sess.run(d, feed_dict=input_dict)
sess.close()
```

```python
%matplotlib inline
a = tf.random_normal([2, 20000], mean=0.0, stddev=2.0)
sess = tf.Session()
out = sess.run(a)
x,y = out

plt.scatter(x, y)
plt.show()
sess.close()
```

```python
a = tf.Variable(0)
init = tf.initialize_all_variables()
s1 = tf.Session()
s2 = tf.Session()

s1.run(init)
s1.run(a.assign_add(5))

s2.run(init)
s2.run(a.assign_add(2))
```
